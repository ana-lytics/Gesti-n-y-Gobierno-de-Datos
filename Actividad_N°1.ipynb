```python
from textblob import TextBlob

# Analizamos el sentimiento de las primeras 10 noticias
for article in articles[:10]:
    title = article['title']
    description = article['description']

    text = f"{title}. {description}" # Fixed indentation

    blob = TextBlob(text) # Fixed indentation

    sentiment = blob.sentiment.polarity  # -1 (negativo) a 1 (positivo) # Fixed indentation

    if sentiment > 0:
        sentiment_label = 'Positive'
    elif sentiment < 0:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    print(f"Title: {title}")
    print(f"Sentiment: {sentiment_label} (Polarity: {sentiment})")
    print('-' * 50)
```

```python
import requests
import time
from datetime import datetime, timedelta

# Lista de ciudades a consultar
cities = ['Philadelphia', 'California', 'Houston', 'North Dakota']

api_key = '996c94a3024245059af234026250404'

# Lista para almacenar los resultados
historical_data = []

def get_historical_weather(city, date):
    url = f'http://api.worldweatheronline.com/premium/v1/past-weather.ashx?key={api_key}&q={city}&date={date}&format=json'

    response = requests.get(url)

    if response.status_code == 200:
        weather_data = response.json()

        # Verificamos si la ciudad es válida
        if 'data' in weather_data and 'weather' in weather_data['data']:
            # Extraemos los datos históricos (temperatura media)
            historical_weather = weather_data['data']['weather'][0]
            avg_temp = historical_weather['avgtempC']  # Temperatura media en Celsius
            weather_description = historical_weather['hourly'][0]['weatherDesc'][0]  # Descripción del clima
            historical_data.append({
                'Date': date,
                'City': city,
                'Temperature': avg_temp,
                'Weather': weather_description
            })
        else:
            print(f"Error: No data available for {city} on {date}")
    else:
        print(f"Error fetching weather data for {city} on {date}: {response.status_code}")

# Generar fechas desde enero 2022 hasta enero 2024 según base de datos que queremos proyectar
start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 1, 1)

# Recorrer las fechas y consultar el clima histórico para cada ciudad
for city in cities:
    print(f"Fetching historical data for {city}...")
    current_date = start_date
    while current_date <= end_date:
        date_str = current_date.strftime('%Y-%m-%d')
        get_historical_weather(city, date_str)

        # Avanzar un día
        current_date += timedelta(days=1)

```

```python
# Lista de ciudades a consultar
cities = ['Philadelphia', 'California', 'Houston', 'North Dakota']

api_key = '5a1cc2a5fd23467cbd5231536250404'

# Función para obtener el pronóstico del clima de una ciudad
# En este caso usé otro API porque superé el límite de calls en el punto anterior
def get_forecast(city):
    url = f'http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days=3'
    response = requests.get(url)

    if response.status_code == 200:
        forecast_data = response.json()

        # Verificamos si la ciudad es válida
        if 'error' in forecast_data:
            print(f"Error fetching forecast data for {city}: {forecast_data['error']['message']}")
        else:
            print(f"Weather forecast for {city}:")
            for day in forecast_data['forecast']['forecastday']:
                date = day['date']
                max_temp = day['day']['maxtemp_c']
                min_temp = day['day']['mintemp_c']
                weather_description = day['day']['condition']['text']
                print(f"{date}: Max Temp: {max_temp}°C, Min Temp: {min_temp}°C, {weather_description}")
            print("------")  # Separador entre ciudades
    else:
        print(f"Error fetching forecast data for {city}: {response.status_code}")

# Recorrer todas las ciudades y obtener su pronóstico
for city in cities:
    get_forecast(city)
```

```python
import requests

api_key = '9f00e4c1a99d692a56addbd4'

# URL base de la API de ExchangeRate-API
base_url = 'https://v6.exchangerate-api.com/v6/'

# Función para obtener el tipo de cambio entre la moneda base y varias monedas
def get_exchange_rates(base_currency, target_currencies):
    url = f'{base_url}{api_key}/latest/{base_currency}'

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        # Verifica si 'conversion_rates' está en los datos de la respuesta
        if 'conversion_rates' in data:
            print(f"Exchange rates for {base_currency}:")
            for target_currency in target_currencies:
                if target_currency in data['conversion_rates']:
                    exchange_rate = data['conversion_rates'][target_currency]
                    print(f"1 {base_currency} = {exchange_rate} {target_currency}")
                else:
                    print(f"Error: No exchange rate found for {target_currency}")
        else:
            print("Error: No conversion rates found in the data.")
    else:
        print(f"Error fetching exchange rates: {response.status_code}")

# Moneda base
base_currency = 'USD'  # US Dollar

# Monedas objetivo
target_currencies = ['EUR', 'PEN', 'GBP', 'CNY', 'CAD']

# Llamar a la función para obtener los tipos de cambio
get_exchange_rates(base_currency, target_currencies)
```

Indicadores de calidad

1. Exactitud (Accuracy)
Qué tan cercanos son los datos extraídos a la realidad. Necesitamos que los datos de temperatura, noticias, y tipo de cambio sean exactos y representen correctamente los eventos y condiciones.

2. Completitud (Completeness)
Verificar que los datos relevantes están presentes y no faltan valores clave. Es fundamental que el conjunto de datos esté completo sin valores nulos en variables clave, como la temperatura, tipo de cambio o noticias relacionadas.

3. Consistencia (Consistency)
Verificar los datos son coherentes.

4. Relevancia (Relevance)
Debemos evaluar si los tipos de cambio, noticias, y temperaturas son variables que realmente tienen impacto en la demanda y precios.

5. Integridad de las Fuentes (Source Integrity)
Debemos verificar la integridad y confiabilidad de las fuentes de datos de las APIs de WeatherAPI y WorldWeatherOnline, NewsAPI y Exchange Rates API.

Estos indicadores son importantes evaluar para verificar la calidad y la relevancia de los datos importados por APIs (Temperatura, Tipo de cambio y Noticias) para pronosticar la demanda y estrategia de precios.
